# 01 - Quickstart

Minimal end-to-end example. Install dependencies with `pip install -r requirements.txt`, copy `configs/config.example.yaml` to `configs/config.yaml`, and set the checkpoint plus input image paths before running the cells below.

```python
from sidewalk_sam.segment import run_segmentation
from sidewalk_sam.utils import load_config, resolve_path

CONFIG_PATH = "configs/config.yaml"
OUTPUT_PATH = "outputs/notebook_roads_mask.png"

config = load_config(CONFIG_PATH)
input_path = resolve_path(config.get("io", {}).get("input_image"))

mask_path = run_segmentation(config_path=CONFIG_PATH, output_path=OUTPUT_PATH)
mask_path
```

```python
from PIL import Image
from IPython.display import display

input_image = Image.open(input_path).convert("RGB")
mask_image = Image.open(mask_path)

display(input_image.resize((512, 512)))
display(mask_image.resize((512, 512)))
```
